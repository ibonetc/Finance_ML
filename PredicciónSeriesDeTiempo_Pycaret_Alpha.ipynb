```python
from pycaret.time_series import *
```

```python
data['Date'] = pd.to_datetime(data['Month'])
```

```python
data.set_index('Date')
```

```python
serie=pd.Series(data['#Passengers'].values, index=data['Date'],name='Number of airline passengers')
```

```python
type(serie)
```

```python
serie
```

```python
s = setup(serie, session_id = 123)
```

```python
import plotly.express as px
fig = px.line(serie,template = 'plotly_dark')
fig.show()
```

```python
check_stats()
```

```python
plot_model(plot = 'decomp_classical')
```

```python
plot_model(plot = 'diagnostics')
```

```python
plot_model(plot = 'acf')
```

```python
plot_model(plot = 'pacf')
```

```python
best = compare_models()
```

```python
best
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh': 24})
```

```python
plot_model(best, plot = 'residuals')
```

```python
plot_model(best, plot = 'insample')
```

```python
final_best = finalize_model(best)
```

```python
save_model(final_best, 'my_best_model')
```

```python
predictions=predict_model(final_best, fh = 48)
predictions
```

```python
serieMonth=pd.Series(data['#Passengers'].values, index=data['Month'])
```

```python
concat_df = pd.concat([serieMonth,predictions], axis=0)
```

```python
concat_df.columns = ['Passengers', 'Prediction']
```

```python
concat_df.index.name='Month'
```

```python
concat_df=concat_df.reset_index()
```

```python
concat_df['Month']=concat_df['Month'].astype('str')
```

```python
fig = px.line(concat_df, x='Month', y=['Passengers', 'Prediction'], template = 'plotly_dark')
fig.show()
```
